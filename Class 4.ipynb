```python
class Matrix:
    def __init__(self, rows, cols):
        self.n_rows = rows
        self.n_cols = cols
        self.rows_list = DoublyLinkedList()

    def append(self, row_list):
        if len(row_list) <= self.n_cols and len(self.rows_list) <= self.n_rows:
            self.rows_list.append(row_list)
        else:
            raise IndexError("Matrix dimensions error")
    def __str__(self):
        out = ""
        for i in range(0,  self.n_rows):
            row_list = self.rows_list.get_value(i)
            for j in range(0, self.n_cols):
                out += str(row_list.get_value(j)) + " "
            out += "\n"
        return out
        
import random
no_rows = 5
no_cols = 9
# matrix = []
matrix = Matrix(no_rows,no_cols)

for numbers in range(no_rows):
    row_list = DoublyLinkedList()
    for number in range(no_cols):
        # row_list.append(number)
        row_list.append(random.randint(10, 99))
    matrix.append(row_list)
print("{} x {} matrix".format(no_rows, no_cols))
print(matrix)

```
